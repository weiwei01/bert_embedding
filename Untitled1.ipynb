```python
from sklearn.decomposition import PCA # Needed for dimension reduction
from bert_serving.client import BertClient
from scipy.spatial import distance
import pandas as pd
from tqdm.notebook import tqdm
from sklearn.cluster import KMeans
bc = BertClient()  # ip address of the GPU machine
vec1 = bc.encode(['我愛吃蘋果'])
vec2 = bc.encode(['我愛蘋果公司的手機'])
```

```python
df1 = pd.read_csv('DataSet2_NoSeg.csv', encoding='utf8')
```

```python
new_df1 = []
for index, row in tqdm(df1.iterrows()):
    print(row['title'])
    tmp_vec = bc.encode([row['title']])
    new_df1.append(tmp_vec[0])
```

```python


kmeans = KMeans(n_clusters=10, max_iter=30)
kmeans.fit(new_df1)# 預測新的目標值 label
new_dy = kmeans.predict(new_df1)
new_dy
```

```python
with open('DataSet2_output.txt', 'w') as f:
    for item in new_dy:
        f.write("%s\n" % item)
```

```python
pca = PCA(n_components=318)
principalComponents = pca.fit_transform(new_df1)
# Plotting the variances for each PC
PC = range(1, pca.n_components_+1)
plt.bar(PC, pca.explained_variance_ratio_, color='gold')
plt.xlabel('Principal Components')
plt.ylabel('Variance %')
plt.xticks(PC)
```

```python
# Putting components in a dataframe for later
PCA_components = pd.DataFrame(principalComponents)
model = KMeans(n_clusters=10,max_iter=30)
model.fit(PCA_components.iloc[:,:2])

labels = model.predict(PCA_components.iloc[:,:2])
plt.scatter(PCA_components[0], PCA_components[1], c=labels)
plt.show()
```
